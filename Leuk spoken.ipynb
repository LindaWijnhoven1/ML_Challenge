```python
#Find features for spoken
print(spoken_train.shape)

def spoken_mean1(signal):
    #Axis=1, is all channel arrays. Output = (N, 13)
    return np.mean(signal, axis=1)
    #..............................
    
def spoken_mean2(signal):
    #Axis=2, is last dimension, namely the time-wise arrays of which we wanna know the mean. Output = (N, 94)
    return np.mean(signal, axis=2)
    #..............................
    
def spoken_length(signal):
    #seeks out the length of a spoken signal. Ignores padding. Required input shape (N,94,13). Returns shape (N,)
    length = signal
    boundary = length < 0
    length[boundary] = 1
    boundary = length > 0
    length[boundary] = 1
    length = np.sum(length, axis=1)
    return np.amax(length,axis=1)
    #..............................

def spoken_distances_13(signal):
    #sums the amplitudes (absolute values) of each characteristic over the entire soundfile. Output shape: (N,13)
    signal_absolute = np.absolute(signal)
    return np.amax(signal_absolute, axis=1)
    #..............................

def spoken_centermass_movement11(signal):
    #identical to spoken_mean1 but with characteristic number 13 and 5 removed across all datapoints. Output: (N,11)
    #why? these are often outliers which would "pull" these hypothetical center of masses away.
    edited_spoken = np.delete(signal,12,axis=2)
    edited_spoken = np.delete(signal,5,axis=2)
    return np.mean(edited_spoken, axis=2)
    #..............................

def spoken_zero_crossover_13(signal):
    #counts the number of sign changes per sound characteristic. Output shape: (N,13)
    signs = np.sign(signal)
    signchange = ((np.roll(signs, 1) - signs) != 0).astype(int)
    return np.sum(signchange, axis=1)


def spoken_MEAN_zero_crossover_13(signal):
    #counts the crossovers of the mean of the 13 characteristics. Shape: (N,)
    step1 = np.mean(signal, axis=1)
    signs = np.sign(step1)
    signchange = ((np.roll(signs, 1) - signs) != 0).astype(int)
    return np.sum(signchange, axis=1)

def spoken_MEAN_zero_crossover_11(signal):
    #counts the crossovers of the mean of the 11 characteristics. The 2 biggest effect having ones have been deleted
    #because those prevent the count of crossovers. Shape: (N,)
    edited_spoken = np.delete(signal,12,axis=2)
    edited_spoken = np.delete(signal,5,axis=2)
    edited_spoken = np.mean(edited_spoken, axis=2)
    signs = np.sign(edited_spoken)
    signchange = ((np.roll(signs, 1) - signs) != 0).astype(int)
    return np.sum(signchange, axis=1)
    
```
